```python
for col in df.columns:
    if df[col].empty:
        continue
    if not np.issubdtype(df[col].dtype,np.number):
        continue
    if np.isinf(df[col]).any() or df[col].isna().any():
        continue
    pearson_coef,p_value = stats.pearsonr(df[col],df['price'])
    print('\nThe Pearson Coefficient between "{}" and price is {}\nwith a p-value of {} !'.format(col,pearson_coef,p_value))
```
