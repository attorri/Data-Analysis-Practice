```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv('Downloads/auto.csv', names = headers)
df.head()
```

```python
import numpy as np

df.replace('?',np.nan, inplace=True)
df.head()
```

```python
null_data = df.isnull()
null_data.head()
```

```python
cols = list(null_data.columns)
for col in cols:
    print('{} : \n{}\n'.format(col,null_data[col].value_counts()))
```

```python
avg_norm_loss = df['normalized-losses'].astype('float').mean(axis=0)
print('Averaged normalized loss - {}'.format(avg_norm_loss))
```

```python
df['normalized-losses'].replace(np.nan, avg_norm_loss,inplace=True)
df.head()
```

```python
avg_bore = df['bore'].astype('float').mean(axis=0)
print('Average bore time - {}'.format(avg_bore))
```

```python
df['bore'].replace(np.nan,avg_bore,inplace=True)
```

```python
avg_stroke = df['stroke'].astype('float').mean(axis=0)
print('Average stroke time {}'.format(avg_stroke))
df.head()
```

```python
df['stroke'].replace(np.nan, avg_stroke, inplace= True)
df.head()
```

```python
df['stroke']
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print('Average horsepower - {}'.format(avg_horsepower))
```

```python
df['horsepower'] = df['horsepower'].replace(np.nan, avg_horsepower)
```

```python
avg_peak_rpm = df['peak-rpm'].astype('float').mean(axis=0)
print('Average peak rpm - {}'.format(avg_peak_rpm))
```

```python
df['peak-rpm'].replace(np.nan,avg_peak_rpm, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
# From above, we can see that four is the mode. 
# And because in number of doors the number must be whole, we'll use the mode
df['num-of-doors'].replace(np.nan, 'four',inplace=True)
df.head()
```

```python
df.dropna(subset=["price"], axis=0,inplace= True)
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df.dtypes
```

```python
df[['bore','stroke']] = df[['bore','stroke']].astype('float')
df[['normalized-losses']] = df[['normalized-losses']].astype('int')
df[['price']] = df[['price']].astype('float')
df[['peak-rpm']] = df[['peak-rpm']].astype('float')
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['city-L/100km'] = 235/df['city-mpg']
df.head()
```

```python
df['highway-L/100km'] = 235/df['highway-mpg']
df.head()
```

```python
df.drop('highway-mpg', axis=1, inplace= True)
df.head()
```

```python
df.head()
```

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```

```python
df['length']
```

```python
df['horsepower'] = df['horsepower'].astype(int, copy=True)
df['horsepower']
```

```python
import matplotlib as plt
from matplotlib import pyplot
```

```python
plt.pyplot.hist(df["horsepower"])
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("Horsepower Bins!")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True)
df['horsepower-binned'].value_counts()
```

```python
pyplot.bar(group_names,df['horsepower-binned'].value_counts())
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower bins')
```

```python
plt.pyplot.hist(df['horsepower'], bins=3)
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower bins')
```

```python
dummy_variable_1 = pd.get_dummies(df['fuel-type'])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'diesel': 'fuel-type:diesel', 'gas': 'fuel-type:gas'},inplace=True)
dummy_variable_1.head()
```

```python
df = pd.concat([df,dummy_variable_1], axis=1)
df.drop('fuel-type', axis=1,inplace=True)
df.head()
```

```python
df.columns
```

```python
dummy_variable_aspiration = pd.get_dummies(df['aspiration'])
dummy_variable_aspiration.head()
```

```python
dummy_variable_aspiration.rename(columns = {'std':'aspiration-std', 'turbo': 'aspiration:turbo'}, inplace=True)
dummy_variable_aspiration.head()
```

```python
df = pd.concat([df,dummy_variable_aspiration], axis=1)
df.drop('aspiration', axis=1, inplace=True)
df.head()
```
